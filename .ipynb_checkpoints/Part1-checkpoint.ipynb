```python
#Classification list
import pandas as pd

classif = pd.read_csv("../FullIJCNN2013/gt.txt", sep=";",header = None, names = ["Name", "left_col", "top_row", "right_col", "bot_row", "class"])

print("The dataset contains ",len(classif)," instances of panel detection")
```

```python
classif.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpl_patches

# Histogram
barhist = {}
for e in classif['class'] :
    if e in barhist :
        barhist[e]+=1
    else :
        barhist[e] = 1

names = list(barhist.keys())
values = list(barhist.values())
color = []
for elt in barhist :
    color.append([int((elt) in Lint), .10,.10])
fig, ax = plt.subplots()
plt.bar(names, values, color=color)
plt.ylabel('Number of instances')
plt.xlabel('Element')


# Legends
labels = []
labels.append("Red bars are the round signs")
labels.append("Black bars are the other signs")


handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 2
ax.legend(handles, labels, loc='best', fontsize='small', 
          fancybox=True, framealpha=0.7, 
          handlelength=0, handletextpad=0)

plt.show()
```

```python
from sklearn.model_selection import train_test_split


Labels = classif[['left_col','top_row','right_col','bot_row']]

Pictures_train, Pictures_test, Label_train, Label_test = train_test_split(classif['Name'], 
                                                                          Labels
                                                                          , test_size=0.2)
```

# Vizualization - Traffic Scene

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np


sample = np.random.randint(0,len(classif),5)

for pic in sample :
    im = np.array(Image.open('../FullIJCNN2013/'+classif['Name'][pic]), dtype=np.uint8)
    fig,ax = plt.subplots(1)
    ax.imshow(im)

    # Label rectangle - Green
    ['left_col','top_row','right_col','bot_row']
    width = classif['top_row'][pic] - classif['bot_row'][pic]
    height = classif['left_col'][pic] - classif['right_col'][pic]    
    rect = patches.Rectangle((classif['right_col'][pic],classif['bot_row'][pic]),height,width,linewidth=1,edgecolor='g',facecolor='none')
    ax.add_patch(rect)
    
    # Output rectangle - Red
    ['left_col','top_row','right_col','bot_row']
    width = classif['top_row'][pic] - classif['bot_row'][pic]
    height = classif['left_col'][pic] - classif['right_col'][pic]    
    rect = patches.Rectangle((classif['right_col'][pic],classif['bot_row'][pic]),height,width,linewidth=1,edgecolor='r',facecolor='none')
    ax.add_patch(rect)
    
    plt.show()
```
